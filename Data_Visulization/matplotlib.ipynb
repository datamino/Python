# **1) Linear Plot**

- Here are all the key attributes that can be used to customize a line plot in matplotlib:

➤**Line making** \
```plt.plot(x, y): Basic plot function with x and y arrays.```

➤**Line Style** \
```linestyle: {'-', '--', '-.', ':'}(Solid, dashed, dashdot, dotted )``` 

➤**Color** \
```color: valid color name or hex code``` 

➤**Width** \
```linewidth: float value in points``` 

➤ **Marker** \
```marker: Marker style like {'.', ',', 'o', 'v', '^', '<', '>' etc}``` 

➤**Marker Size** \
```markersize: float size in points``` 

➤**Marker Color** \
```markerfacecolor: color of markers``` 
 
```markeredgecolor: color of markers``` 

````python
# Two Numpy Arrays

x_axis=np.array([1,2,3,4,5,6])
y_axis=np.array([5,9,2,8,1,2])

# Linear Plot in Matplotlib.
plt.plot(x_axis,y_axis)
plt.title('Linear Plot')  
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
````

````python
# Linear Plot in Matplotlib
plt.plot(x_axis,y_axis,linestyle='-',color='Black',markerfacecolor='r',markeredgecolor='red',marker='o',markersize=8,linewidth=2 )
plt.title('Linear Plot')  
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
````
